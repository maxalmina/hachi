# Particle Swarm Optimization
---
In computational science, particle swarm optimization (PSO) is a computational method that optimizes a problem by iteratively trying to improve a candidate solution with regard to a given measure of quality. It solves a problem by having a population of candidate solutions, here dubbed particles, and moving these particles around in the search-space according to simple mathematical formulae over the particle's position and velocity. Each particle's movement is influenced by its local best known position, but is also guided toward the best known positions in the search-space, which are updated as better positions are found by other particles. This is expected to move the swarm toward the best solutions.

## Import All Necessary Tools

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

## The Functions
In this notebook we'll perform analysis, design, and implementation of Particle Swarm Optimization (PSO) into a code that is able to find the minimum value of two functions below:

$$f(x_1, x_2) = - \sum_{i=1}^{5} i \cos{((i+1)x_1 + 1)} \sum_{i=1}^{5} i \cos{((i+1)x_2 + 1)}$$

$$f(x_1, x_2) = -\cos(x_1)\cos(x_2)exp(-(x_1-\pi)^2-(x_2-\pi)^2)$$

where $-100 \leq x_1 \leq 100$ and $-100 \leq x_2 \leq 100$. Let's create the functions above in python syntax

```python
def fungsi1(x):
    return (-1 * np.sum(i*np.cos((i+1)*x[0]+1) for i in range(1,6)) * np.sum(i*np.cos((i+1)*x[1]+1) for i in range(1,6)))

def fungsi2(x):
    return -np.cos(x[0])*np.cos(x[1])*np.exp(-(x[0]-np.pi)**2 - (x[1]-np.pi)**2)
```

## Visualize The Functions
We're trying to describe the functions through visualization below:

```python
def visualisasi(fungsi):
    x = np.linspace(-100, 100, 300)
    y = np.linspace(-100, 100, 300)

    X, Y = np.meshgrid(x, y)
    Z = fungsi([X, Y])

    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.contour3D(X, Y, Z, 50)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.set_title('3D contour')
    plt.show()
```

## Fitness Function
The functions above if we visualize, will be returned either positive or negative value. So, for the fitness function we used this formula $fitness = c^{-h}$ where $c=2$.

```python
def fitness1(x):
    return 2**(-fungsi1(x))

def fitness2(x):
    return 2**(-fungsi2(x))
```

## Helper Functions

Helper functions below will help our algorithm. First we create euclidean function to calculate distance between two fireflies using their coordinates.

```python
def euclidean(a, b):
    sm = ((a[0] - b[0])**2) + ((a[1] - b[1])**2)
    return np.sqrt(sm)
```

The FA individuals movements are influenced by attractiveness of other FA individuals. The attractivenss formulated by this formula:
$$\beta(r) = \beta_0 e^{-\gamma r_{ij}^2}$$
where r is distance between two fireflies and $\gamma$ is light absorbtion coefficient. The value of $\beta_0$ will be 0 if the value of r also 0.

```python
def attractiveness(beta0, gamma, r):
    if r == 0:
        beta0 = 0
    return beta0*np.exp(-gamma*(r**2))
```

## PSO Algorithm
Let's design the PSO Algorithm. Before that, we must determine PSO parameters.


```python
def PSO(fitness, fungsi, beta0, upper, lower, n_population=40, max_gen=20):
    L = upper - lower
    gamma = 0.5/(L**2)
    alpha = 0.01*L

    # creating empty numpy.array contains position of firefly and its fitness
    population = np.zeros(shape=(n_population, 3))

    for i in range(n_population):
        x1 = random.uniform(-100,100)
        x2 = random.uniform(-100,100)
        I = fitness([x1,x2])
        population[i] = ([x1, x2, I])

    # initialize best firefly 
    best_ind_index = np.argmax(population[:,-1])
    best_ind = population[best_ind_index]

    for i in range(max_gen):
        for j in range(len(population)):
            b = True # tidak ada individu yang lebih baik dari j diset True
            for k in range(len(population)):
                # attractiveness
                r = euclidean(population[j,:-1], population[k,:-1])
                attr = attractiveness(beta0, gamma, r)

                if population[k,-1] > population[j,-1] :
                    b = False
                    population[j,0] += (attr*(population[k,0]-population[j,0])) + (alpha*(random.uniform(0,1) - 0.5))
                    population[j,1] += (attr*(population[k,1]-population[j,1])) + (alpha*(random.uniform(0,1) - 0.5))
                    population[j,-1] = fitness(population[j,:-1])

            # jika tidak ada individu k yang lebih baik dari individu j
            if b == True:
                population[j,0] = random.uniform(-100,100)
                population[j,1] = random.uniform(-100,100)
                population[j,-1] = fitness(population[j,:-1])
      
        # mengambil solusi terbaik sementara
        best_ind_index_temp = np.argmax(population[:,-1])
        best_ind_temp = population[best_ind_index_temp]
        if best_ind_temp[-1] > best_ind[-1]:
            best_ind = best_ind_temp
    
    # decrease alpha
    alpha *= 0.97**i

    # solusi FA pada kasus minimasi fungsi
    return best_ind[:-1], fungsi(best_ind)
```
